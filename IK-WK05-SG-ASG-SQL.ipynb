Setup
**Install SQLite **
If you're on a Linux or Mac machine, you should already have SQLite. To test, run:
sqlite3 --version
If you need to install it, choose the correct precompiled binary for your operating system on the SQLite Download Page.

**Setup the Tickit Database**
The Tickit Database is a sample database published by Amazon Web Services.
Download the tickit.sql file & tickitdb directory from this repository.
Create a directory called "sql_assignment" on your local drive and move the above-downloaded files into the directory.
Next open the Terminal or Command Window and type "cd sql_assignment" to go into the directory you just created.
Open the SQLite Command Line Shell by running sqlite3 on Mac or Linux or sqlite3.exe on Windows from the directory you've saved tickit.sql file & tickitdb. You should see output similar to:
SQLite version 3.28.0 2019-04-15 14:49:49
Enter ".help" for usage hints.
Connected to a transient in-memory database.
Use ".open FILENAME" to reopen on a persistent database.
sqlite>
Run the command .open tickit.db to make your database persistent. This will create a file tickit.db on your local machine that stores your database.
Run the command .read tickit.sql to set up the Tickit Database.
Run the command .tables to confirm everything is set up. You should see output similar to:
sqlite> .tables
category  date      event     listing   sales     users     venue
Run the command .quit to quit the database.
When you return to do more SQL problems you'll only need to do steps 2 and 3. You can also run sqlite3 tickit.db if you're on a Mac or Linux machine.

**SQLite Tips**
.schema TABLE_NAME can be used to look up the schema for a table.
.read solutions/question_1.sql will run the SQL in the file solutions/question_1.sql on your local machine.
