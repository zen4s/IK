Numpy-FF4:
Using NumPy, create a 1000-element array of random integers between 1 and 1000. Calculate the cumulative sum of the array and print the 10th, 100th, and 500th elements of the cumulative sum array.

```python
ref1000 = list(np.arange(1,1001))     #Ordered list with 1 to 1000

np_a4 = np.random.choice(ref1000, size=(1000,))   #1000-element array of random integers between 1 and 1000.
print(np_a4)

#Option #1
cum_sum = np.cumsum(np_a4) #does in place update

#Option #2

list1000 = []
list1000 = list(np_a4)

total = 0
for i,val in enumerate(list1000):
    total += val
    if i in [9, 99, 499]:
        print (f"The element# {i+1} is {val} & Cummulative is {total} ")

print(f"10th= {cum_sum[9]}, 100th= {cum_sum[99]}, 500th= {cum_sum[499]}, Array Total= {total}")
```

-FF10:

Write a Python program using NumPy to create a 2D array of size 8x8, where the outer border elements are set to 1 and the inner elements are set to 0. Print the resulting array.

```python
row_size = 8
col_size = 8
sizea = row_size*col_size
np_a9 = np.zeros((sizea,), dtype = int)   #. create a flat zero array
row_edge = np.ones((col_size,), dtype = int)   # create a flat one's array with size of one row

print(np_a9)
print(row_edge)

#Changes all 1st Col & 8th col through the loop
for i in range(0, len(np_a9), col_size):          #change iterate the array and pick 1st & 8th (col_size) elements --> [0] & [7] and increment by 8 (col_size)
    np_a9[i] = np_a9[i+col_size-1] = 1

np_a9 = np_a9.reshape(row_size, col_size)        #reshape to matrix 8x8
print(np_a9)

np_a9[0] = np_a9[row_size - 1] = row_edge       #set the top and bottom row edge

print(np_a9)
```
