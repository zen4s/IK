```python
import numpy as np

step_n = 10
steps = np.empty([step_n,2]) #emply numpy array of 10 one element list
print("Empty Numpy Array....")
print(steps)

steps = np.random.choice([-1, 0, 1], size=(1,2)) # from a list of [-1,0,1] create a numpy array of 1x2 shape, with random selection
print("Steps before loop.....")
print(steps)

for n in range(step_n-1):
    step = np.random.choice([-1, 0, 1], size=(1,2))
    print(f"Step for count #{n} is ....")
    print(step)

    steps = np.append(steps, step, axis=0)
    #something will be checked after each n
    print("steps.. after append ...")
    print(steps)

print("Steps .....")
print(steps)
```
