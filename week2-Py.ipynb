```python
haystack = "sadbutsad"
needle = "sads"
nl,hl = len(needle),len(haystack)
print(nl,hl)
if nl == 0:
  print(0)
elif hl < nl :
  print(-1)
else :
  for i in range(hl - nl + 1) :
    if haystack[i:i+nl] == needle:
      print(i)
  print(i)
  print(hl - nl + 1)
  if i > (hl - nl + 1) :
    print(-1)
```
