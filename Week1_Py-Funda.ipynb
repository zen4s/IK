```python
import math
max_subarray = -math.inf
for i in range(len(numbers)):
    current_subarray = 0
    for j in range(i, len(numbers)):
        current_subarray += numbers[j]
        max_subarray = max(max_subarray, current_subarray)
print(max_subarray)
```
