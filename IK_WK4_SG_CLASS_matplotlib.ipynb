```python
# load libarary
# Import the pyplot module from the matplotlib library
import matplotlib.pyplot as plt
```

```python
# Sample data: Product categories and revenue
categories = ['Electronics', 'Clothing', 'Accessories', 'Home',
'Beauty' ]
revenue = [40000, 25000, 18000, 35000, 22000]
```

```python
plt.bar(categories,revenue)
plt.title("Revenue by Categoty")
plt.xlabel("Product Categories")
plt.ylabel("Revenue ($)")

for i, rev in enumerate(revenue):
  plt.text(i,
           rev,
           rev,
           ha='center',
           va='bottom',
           fontsize=10
           )

plt.show()
```
